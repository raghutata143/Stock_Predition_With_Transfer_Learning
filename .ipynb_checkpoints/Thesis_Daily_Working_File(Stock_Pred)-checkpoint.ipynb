```python
## Test Stationarity for time series data

def test_stationarity_plot(timeseries):
    
    #Determing rolling statistics
    rolmean = pd.rolling_mean(timeseries, window=12)
    rolstd = pd.rolling_std(timeseries, window=12)

    #Plot rolling statistics:
    orig = plt.plot(timeseries, color='blue',label='Original')
    mean = plt.plot(rolmean, color='red', label='Rolling Mean')
    std = plt.plot(rolstd, color='black', label = 'Rolling Std')
    plt.legend(loc='best')
    plt.title('Rolling Mean & Standard Deviation')
    plt.show(block=False)
    
#     #Perform Dickey-Fuller test:
#     print('Results of Dickey-Fuller Test:')
#     dftest = adfuller(timeseries, autolag='AIC')
#     dfoutput = pd.Series(dftest[0:4], index=['Test Statistic','p-value','#Lags Used','Number of Observations Used'])
#     for key,value in dftest[4].items():
#         dfoutput['Critical Value (%s)'%key] = value
#     print(dfoutput)
```

```python
## Data preprocessing generate labels (buy,sell,hold) writing the function that gonna map to pandas dataframe new column for 
# the target (buy,sell and hold) based on percentage error of the price


# By below heatmap there is definitely correlation b/w companies pricing data (cerain companies, group of companies etc..)
# My question is can machine recognize those relations.
# My thesis is group of companies can move together up / down but its not necessary that they exactly at the same time. 
# Someone's gonna be a
# 1st moves and some may be last. Can we take lot of time people look at one comapny pricing history for a time period 
# and predict howz the price gonna be.
# My focus of work is by using pricing data can we make decisions such as (buy, sell, hold)
#  percentage change of price data is used as features so that we can decide option about stock (buy, sell, data).
# we gonna take all the features data and the way we gonna generate buy,sell, hold based on training data. 
# We gonna ask question on training data did price based on these features right here within next 7 trading days did
# price go up more than 2% then buy, if it falls more than 2% then sell, neither of it is true then hold the stock.


## Abbreviations in the below code:
# hm_days = how many days in the future stock price to make or loose x%
# df['{}_{}d'.format(ticker, i)] = creating a custom column (ticker,i) = day i in future what is the value of the price
#                                  and we are say that in percentage change.percentage change is calculated by 
#                                  (price in i days from now - today's price / today's price times 100)

# (df[ticker].shift(-i) - df[ticker]) / df[ticker] = shift(-i) means shifting things up / down. In this case we are shiting 
#                                                    -vely because we wanted future data. When we shift negatively we are 
#                                                    shifting up, because we are getting future data. 
#                                                    What is this is gonna do is literally if we consider data with two hands
#                                                    left most column and date in rows, future date is at the bottom, 
#                                                    so we cans shift negatively means we are moving up and getting future data

# basically classification each model we gonna generate per company basis but each company is going to take into account 
# pricing data of all the companies
# ---------------------------------------------------------------------------------------------------------------------------


# function defining whether to buy, sell and hold the stock
# return 1 = Buy, retrun -1 = Sell, 0 = Hold
# setting our percentage change of price. Take all the features normalised price values and we gonna ask the question on training
# data whether buy, sell or hold Did the price based on these features right here within next 7 trading days did the price goes
# up 2% then buy, did the price goes down by 3% then sell, if neither of the case is true then Hold)


def buy_sell_hold(*args): 
    cols = [c for c in args] # getting value of price for particular stock from joined closing csv file
    Buy = 1
    Sell = -1
    Hold = 0
    
#   (0.03 and -0.10 are % change of price if prices goes up by 30% and if price goes down by -10%)
    for col in cols:
        if col > 0.03:
            return Buy
        if col < -0.10:
            return Sell      
    
    
    return Hold # giving return 0 returns to hold the stock. But its not required because we are interested in only 
                # buying / selling the stock for profits 



## Preparing data for classifier labels Buy, Sell,or Hold

hm_days = 7
def process_data_for_labels(ticker): 
    df = pd.read_csv('sp500_joined_closing_prices.csv', index_col = 0)
    tickers = df.columns.values.tolist() # Ticker names
    df.fillna(0, inplace =  True) # remove 0 values
    
    for i in range(1, hm_days+1): # (range of 7 days from day 1 through 7)
        df['{}_{}d'.format(ticker, i)] = (df[ticker].shift(-i) - df[ticker]) / df[ticker] # (new - old / ticker)
        
    df.fillna(0, inplace = True)
    return tickers, df




# Mapping the above functions to a Dataframe created 'Price column'
# Now the below function is to extract featuresets

def extract_featuresets_z_norm(ticker):
    tickers, df = process_data_for_labels(ticker)
    
    # The below code passes 7 days price percentage changes for the future of the stock
    df['{}_target'.format(ticker)] = list(map( buy_sell_hold,
                                             df['{}_1d'.format(ticker)],
                                             df['{}_2d'.format(ticker)],
                                             df['{}_3d'.format(ticker)],
                                             df['{}_4d'.format(ticker)],
                                             df['{}_5d'.format(ticker)],
                                             df['{}_6d'.format(ticker)],
                                             df['{}_7d'.format(ticker)]))
    
    vals = df['{}_target'.format(ticker)].values.tolist()
    str_vals = [str(i) for i in vals]
    print('Spread of ' + str(hm_days)+ ' days price percentage changes for ' + ticker + ':\t', Counter(str_vals))
    
    ## Normalising percentage change closing price data of all the stocks for plotting -- z-normalization
    
    # Creating feature_sets and labels. Here ticker is just prices and normalised with z-score normalization and returning the 
    # normalized values (df_vals) for plotting purposes
    
    df2 = df[[ticker for ticker in tickers]]
    percentage_price_change_of_stocks = df2.apply(zscore)
    percentage_price_change_of_stocks = pd.DataFrame(percentage_price_change_of_stocks)
    percentage_price_change_of_stocks.fillna(0, inplace = True)
    percentage_price_change_of_stocks = percentage_price_change_of_stocks.replace([np.inf, -np.inf], 0)
    percentage_price_change_of_stocks.fillna(0, inplace = True)
    
    # Features(X = array of futute price Percentage changes for all the companies including the company), 
    # y = Targets(buy,sell,hold)
    
    X = percentage_price_change_of_stocks.values
    y = df['{}_target'.format(ticker)].values
    
    return X, y, df 


def do_ml_z_norm(ticker):
    X, y, df = extract_featuresets_z_norm(ticker) 
    
    X_train, X_test, y_train, y_test = cross_validation.train_test_split(X,
                                                                         y,
                                                                         test_size = 0.25)
    
    clf = VotingClassifier([('SVM Classifier', svm.LinearSVC()),
                           ('KNN Classifier', neighbors.KNeighborsClassifier(n_neighbors=10, weights='uniform')),
                           ('Random Forest Classifier', RandomForestClassifier(max_depth=2, random_state=0))])
    
    clf.fit(X_train, y_train)
    predictions = clf.predict(X_test)
    print('\nPredictions Spread for ' + ticker + ' in order Buy, Sell and Hold:\t', Counter(predictions))
    confidence = print('\nAccuracy of the model:\n',clf.score(X_test, y_test))
    
    return confidence


do_ml_z_norm('GOOG')

```
